```python
keywords = ["Helping"]

def scan_for_articles(keyword):
    newsapi = NewsApiClient(api_key=NEWS_API_KEY)
    relevant_articles = newsapi.get_everything(q=keyword,
                                        language='en', 
                                        sort_by='relevancy', 
                                        page_size=1)

    return relevant_articles


#relevant_articles_df = pd.DataFrame(scan_for_articles()['articles'])
```

```python
x = scan_for_articles(keywords[0])
x_df = pd.DataFrame.from_dict(x['articles'])
x_df["source"] = x_df["source"].apply(lambda x: x['name'])


```

```python
def make_clickable(url):
    return f'<a target="_blank" href="{url}">{url}</a>'

x_df.style.format({'url': make_clickable})
```
