```python
from datetime import date, timedelta
import pandas as pd
import os
import csv
import numpy as np
from pathlib import Path
from dotenv import load_dotenv
load_dotenv()
```

```python
file_path = Path("all_relevant_articles.csv")
relevant_element_data = pd.read_csv(file_path)
relevant_element_data.head()
```

```python
over_90_confidence_df = relevant_element_data[relevant_element_data.confidence > 0.9]
over_90_confidence_df.head()
```

```python
over_90_confidence_df_sorted = over_90_confidence_df.sort_values('sentiment')
over_90_confidence_df_sorted
```

```python
over_90_confidence_positive = over_90_confidence_df_sorted[over_90_confidence_df_sorted['sentiment'] == "POSITIVE"]
over_90_confidence_positive.head()
```

```python
over_90_confidence_negative = over_90_confidence_df_sorted[over_90_confidence_df_sorted['sentiment'] == "NEGATIVE"]
over_90_confidence_negative.head()
```
